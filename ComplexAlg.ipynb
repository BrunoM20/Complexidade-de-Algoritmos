> ## Cálculo do tempo de processamento (em segundos).
---
+ ### Criação de uma lista com 1 milhão de valores aleatórios entre 1 e 1000 e, em seguida, 3 buscas para encontrar a qtde de valores iguais a 500.

```python
from random import randint
from time import time

MAX = 10 ** 6

t = time()
dados = [randint(1, 1000) for i in range(MAX)]
dt = time() - t
print(f'Tempo gasto na criação da lista = {dt:.3f} s')

t = time()
q500 = 0
for i in range(MAX):
    if dados[i] == 500:
        q500 += 1
dt = time() - t
print(f'\nQtde iguais a 500 (por indexação) = {q500}', end='   ')
print(f'Tempo gasto = {dt:.3f} s')

t = time()
q500 = 0
for v in dados:
    if v == 500:
        q500 += 1
dt = time() - t
print(f'\nQtde iguais a 500 (acesso direto pelo for) = {q500}', end='   ')
print(f'Tempo gasto = {dt:.3f} s')

t = time()
q500 = dados.count(500)
dt = time() - t
print(f'\nQtde iguais a 500 (através do método count) = {q500}', end='   ')
print(f'Tempo gasto = {dt:.3f} s')
```

> ## Analisar a complexidade:
---
+ ## f(n) = n² + 1000n + 200000 ----> Complexidade O(n²)
+ ## Comportamento dos termos da equação em relação ao valor de n (entrada).

```python
from time import sleep
from IPython.display import clear_output

for n in range(1, 1000000):
    t1 = n ** 2
    t2 = 1000 * n
    t3 = 200000
    print(f'n: {n}:   n²: {t1}   1000n: {t2}   {t3}')
    print(f'\nProporção 1000n por 200000: {1000 * n / 200000:.2f}')
    print(f'\nProporção n² por 1000n: {n ** 2 / (1000 * n):.2f}')
    print('\nMaior peso: ', end='')
    if t3 > t1 and t3 > t2:
        print('200000')
    elif t2 > t1:
        print('1000n')
    else:
        print('n²')
    sleep(0.05)
    clear_output(True)
```

> ## Análise da complexidade de tempo (gráfico):
---
 + ### O(n²)
 + ### O(n³)
 + ### O(n log2 n)
 + ### O( log2 n)

```python
import matplotlib.pyplot as plt
from ipywidgets import interact
import math

eixo_x = range(1, 1000)  # tamanho da entrada (n)
y1 = [n ** 3 for n in eixo_x]  # O(n³)
y2 = [n ** 2 for n in eixo_x]  # O(n²)
y3 = [n * math.log2(n) for n in eixo_x] # O(n log n)
y4 = [n for n in eixo_x]  # O(n)
y5 = [math.log2(n) for n in eixo_x]  # O(log n)

@interact(amp = (1, 1000))
def fx(amp = 10):
    plt.figure(figsize = (5, 4))
    plt.plot(eixo_x, y1) # O(n³)
    plt.plot(eixo_x, y2) # O(n²)
    plt.plot(eixo_x, y3) # O(n log n)
    plt.plot(eixo_x, y4)  # O(n)
    plt.plot(eixo_x, y5) # O(log n)
    plt.xlim(0, amp)
    plt.ylim(0, amp ** 2)
    plt.show()
```

> ## Análise da Complexidade (tempo) de Algoritmos de Busca (Pesquisa):
---
 + ### Busca sequencial
 + ### Busca binária (iterativa)
 + ### Busca binária (recursiva)

> ### **Busca sequencial - Complexidade O(n):**

```python
def busca_seq(dados, valor):
  for i in range(len(dados)):
    if dados[i] == valor:
      return i
  return -1
```

> ### **Busca binária (iterativa) - Complexidade O(log n):**

```python
def busca_bin_ite(dados, valor):
  i, f = 0, len(dados) - 1
  while i <= f: # repete enquanto o início não ultrapassar o fim
    ic = (i + f) // 2 # calcula o índice central (quociente da divisão por 2)
    if valor == dados[ic]: # verifica se o valor é igual ao elemento central
      return ic  # valor encontrado
    elif valor < dados[ic]: # verifica se a chave é menor que o elemento central
      f = ic - 1
    else:   # o valor é maior que o elemento central
      i = ic + 1
  return -1 # valor não encontrado
```

> ### **Busca binária (recursiva) - Complexidade O(log n):**

```python
def busca_bin_rec(dados, valor, i, f):
  if i <= f: # verifica se o início ainda não ultrapassou o fim
    ic = (i + f) // 2 # calcula o índice central (quociente da divisão por 2)
    if valor == dados[ic]: # verifica se o valor é igual ao elemento central
      return ic  # valor encontrado
    elif valor < dados[ic]:
      return busca_bin_rec(dados, valor, i, ic - 1) # chama a função novamente alterando o final
    else:
      return busca_bin_rec(dados, valor, ic + 1, f) # chama a função novamente alterando o início
  else:
    return -1 # valor não encontrado
```

```python
# Programa Principal

from random import randint, sample
from time import time

lista = sample(range(10000000), 10000000)
```

```python
v = int(input('Valor a ser encontrado: '))

# Busca Sequencial
t = time()
pos = busca_seq(lista, v)
dt = time() - t
print(f'\nSequencial: Posição = {pos}  Busca: {dt:.3f} s')

# Busca binária - ordenar lista
to = time()
lista.sort() # Tim sort
dto = time() - to

# Busca binária iterativa
t = time()
pos = busca_bin_ite(lista, v)
dt = time() - t
print(f'\nBinária iterativa: Posição = {pos}  Ordenação: {dto:.5f} s + Busca: {dt:.5f} s = {dt + dto:.5f} s')

# Busca binária recursiva
t = time()
pos = busca_bin_rec(lista, v, 0, len(lista) - 1)
dt = time() - t
print(f'\nBinária recursiva: Posição = {pos}  Ordenação: {dto:.5f} s + Busca: {dt:.5f} s = {dt + dto:.5f} s')
```

###Bubble Sort Otimizado:

```python
def bubbleSO(lis):
  n = len(lis)
  for i in range(n):
    troca = False
    for j in range(0, n - i - 1):
      if lis[j] > lis[j + 1]:
        # Troca os elementos
        lis[j], lis[j + 1] = lis[j + 1], lis[j]
        troca = True
      # Se não houve troca, a lista já está ordenado
    if not troca:
      break
```

###Selection Sort:


```python
def selectionS(lis):
  n = len(lis)
  for i in range(n):
    min_ind = i
    for j in range(i + 1, n):
      if lis[j] < lis[min_ind]:
        min_ind = j
    lis[i], lis[min_ind] = lis[min_ind], lis[i]
```

###Quick Sort:

```python
def quickS(lis):
  if len(lis) <= 1:
    return lis
  else:
    pivo = lis[len(lis) // 2]
    left = [x for x in lis if x < pivo]
    mid = [x for x in lis if x == pivo]
    right = [x for x in lis if x > pivo]
    return quickS(left) + mid + quickS(right)
```

###Merge Sort:

```python
def mergeS(lis):
  if len(lis) > 1:
    mid = len(lis) // 2
    left = lis[:mid]
    right = lis[mid:]

    mergeS(left)
    mergeS(right)

    i = j = k = 0

    while i < len(left) and j < len(right):
      if left[i] < right[j]:
        lis[k] = left[i]
        i += 1
      else:
        lis[k] = right[j]
        j += 1
      k += 1

    while i < len(left):
      lis[k] = left[i]
      i += 1
      k += 1

    while j < len(right):
      lis[k] = right[j]
      j += 1
      k += 1
```

###Tim Sort:

```python
def timS(lis):
  lis.sort()
```
